```python
import torch
import torch.nn as nn
import torch.nn.functional as F # Import F for entropy and other potential functions

# Content of VulnerabilitySpotter class
vulnerability_spotter_code = """
import torch
import torch.nn as nn
import torch.nn.functional as F # Import F for entropy and other potential functions

class VulnerabilitySpotter(nn.Module):
    \"\"\"
    Vulnerability Spotter module for detecting multi-metrics (scarcity, entropy, deceptive risk)
    and aggregating them into a v_t score using log-odds.
    \"\"\"
    def __init__(self, d_model: int = 256):
        \"\"\"
        Initializes the VulnerabilitySpotter with layers for detecting specific metrics.

        Args:
            d_model: The dimensionality of the input and output features.
        \"\"\"
        super().__init__()
        self.d_model = d_model

        # Layers for Scarcity Detection (Example: Focus on specific feature patterns)
        # This is a placeholder. Real scarcity detection might involve learned patterns
        # or external knowledge.
        self.scarcity_linear = nn.Linear(d_model, 1)

        # Layers for Entropy Detection (Example: Focus on distribution or variance)
        # This is a placeholder. Real entropy detection might involve analyzing activation
        # distributions or using specific attention mechanisms.
        self.entropy_linear = nn.Linear(d_model, 1) # Simplified: predict entropy-like score

        # Layers for Deceptive Risk Detection (Example: Focus on hidden variance or inconsistencies)
        # This is a placeholder. Real deceptive risk might involve analyzing hidden states
        # or comparing with known deceptive patterns.
        self.deceptive_risk_linear = nn.Linear(d_model, 1)


        # Layer for aggregating metrics before log-odds (Optional, can go directly to log-odds)
        # self.aggregation_linear = nn.Linear(3, 1) # Aggregate scarcity, entropy, deceptive risk scores

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        \"\"\"
        Forward pass of the VulnerabilitySpotter to detect metrics and compute v_t.

        Args:
            x: Input tensor of shape (batch_size, sequence_length, d_model).

        Returns:
            Output tensor representing the aggregated vulnerability score (v_t)
            of shape (batch_size, sequence_length, 1) or (batch_size, 1).
        \"\"\"
        batch_size, seq_len, d_model = x.size()

        # Flatten the input for linear layers if processing token-wise
        x_flat = x.view(-1, d_model) # Shape: (batch_size * sequence_length, d_model)

        # Detect individual metrics (placeholder implementations)
        scarcity_scores = torch.sigmoid(self.scarcity_linear(x_flat)) # Shape: (batch_size * seq_len, 1)
        # Example of using variance for entropy (simplified, not true entropy)
        # You'd need a more sophisticated approach for real entropy.
        entropy_scores = torch.sigmoid(self.entropy_linear(x_flat)) # Shape: (batch_size * seq_len, 1)
        deceptive_risk_scores = torch.sigmoid(self.deceptive_risk_linear(x_flat)) # Shape: (batch_size * seq_len, 1)


        # Concatenate the metric scores
        # Shape: (batch_size * sequence_length, 3)
        metric_scores = torch.cat((scarcity_scores, entropy_scores, deceptive_risk_scores), dim=-1)

        # Aggregate metrics and apply log-odds (Simplified example)
        # The paper's log-odds aggregation involves probabilities. Assuming metric_scores
        # can be treated as probabilities or values proportional to them.
        # A simple aggregation before log-odds:
        aggregated_score = torch.mean(metric_scores, dim=-1, keepdim=True) # Shape: (batch_size * seq_len, 1)

        # Convert score to a value between epsilon and 1-epsilon for stable log-odds
        epsilon = 1e-6
        p = torch.clamp(aggregated_score, epsilon, 1.0 - epsilon)

        # Calculate log-odds: log(p / (1-p))
        log_odds = torch.log(p / (1 - p))

        # Aggregate log-odds across the sequence length (example: mean)
        # The paper's aggregation might be more complex (e.g., weighted sum, max).
        # Here, we reshape back to (batch_size, seq_len, 1) before aggregating along seq_len
        log_odds = log_odds.view(batch_size, seq_len, 1)

        # Returning token-level log_odds for now, reshape to (batch_size, seq_len, 1).
        # If a single score per sequence is needed, aggregate further (e.g., mean(dim=1))
        v_t = log_odds # Keep as token-level scores

        return v_t # Shape: (batch_size, sequence_length, 1)
"""

# Overwrite the existing file or create it
with open("cal.py", "w") as f:
    f.write(vulnerability_spotter_code)

print("VulnerabilitySpotter class written to cal.py")
```

### Subtask:

```python
import torch.nn as nn
import torch

# Content of ConfessionalTemplate class
confessional_template_code = """
import torch.nn as nn
import torch

class ConfessionalTemplate(nn.Module):
    \"\"\"
    Confessional Template module for generating different confessional outputs
    using named templates.
    \"\"\"
    TEMPLATES = {
        'prior': "What assumptions am I making? What is the context? What is the goal?",
        'evidence': "What do I observe? What data is relevant? What are the facts?",
        'posterior': "What is the most likely explanation? What are the implications?",
        'relational': "How does this relate to other information? Are there inconsistencies?",
        'moral': "What are the ethical considerations? What are the potential harms?",
        'action': "What should I do next? What is the safest course of action?"
    }

    def __init__(self, d_model: int = 256):
        \"\"\"
        Initializes the ConfessionalTemplate with named templates.

        Args:
            d_model: The dimensionality of the input and output features.
        \"\"\"
        super().__init__()
        self.d_model = d_model
        self.template_projections = nn.ModuleDict({k: nn.Linear(d_model, d_model) for k in self.TEMPLATES})

    def structure_reasoning(self, z: torch.Tensor, step: str = 'prior') -> torch.Tensor:
        \"\"\"
        Applies a template projection to the input tensor.

        Args:
            z: Input tensor representing the private thought state (batch_size, sequence_length, d_model).
            step: The name of the template to use (e.g., 'prior').

        Returns:
            Output tensor after applying the selected template projection.
        \"\"\"
        if step in self.template_projections:
            # Add a small amount of noise as suggested in the patch snippet
            return self.template_projections[step](z) + torch.randn_like(z) * 0.01
        # If the step is not a recognized template, return the input unchanged
        return z

    def forward(self, z: torch.Tensor, step: str = 'prior') -> torch.Tensor:
        \"\"\"
        Forward pass of the ConfessionalTemplate.

        Args:
            z: Input tensor representing the private thought state (batch_size, sequence_length, d_model).
            step: The name of the template to use (e.g., 'prior').

        Returns:
            Output tensor after applying the selected template.
        \"\"\"
        return self.structure_reasoning(z, step)
"""

# Append the class definition to cal.py
with open("cal.py", "a") as f:
    f.write("\n") # Add a newline for separation
    f.write(confessional_template_code)

print("ConfessionalTemplate class added to cal.py")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Content of TinyConfessionalLayer class
tiny_confessional_layer_code = """
import torch
import torch.nn as nn
import torch.nn.functional as F

class TinyConfessionalLayer(nn.Module):
    \"\"\"
    Tiny Confessional Layer with a recursive think/act loop and coherence computation.
    \"\"\"
    def __init__(self, d_model: int = 256, num_iterations: int = 16, n_inner: int = 6, audit_mode: bool = False): # Increased num_iterations, added n_inner and audit_mode
        \"\"\"
        Generates a new notebook cell with the TinyConfessionalLayer class definition and appends it to cal.py.
        \"\"\"
        super().__init__()
        self.d_model = d_model
        self.num_iterations = num_iterations
        self.n_inner = n_inner
        self.audit_mode = audit_mode

        self.vulnerability_spotter = VulnerabilitySpotter(d_model)
        self.confessional_template = ConfessionalTemplate(d_model)

        # Layers for the 'think' step
        # Think input will be concatenation of x_pooled, y, and z
        self.think_linear1 = nn.Linear(d_model * 3, d_model) # Adjusted for concatenated input
        self.think_relu = nn.ReLU()
        self.think_linear2 = nn.Linear(d_model, d_model)

        # Layers for the 'act' step
        # Act input will be concatenation of y and z
        self.act_linear = nn.Linear(d_model * 2, d_model) # Added for act step


        # Layers for coherence computation (cosine similarity + KL divergence)
        # Cosine similarity layer (compares two vectors of d_model)
        self.cosine_sim = nn.CosineSimilarity(dim=-1)

        # KL divergence requires probability distributions. Assuming we can get
        # distributions from the states (e.g., via softmax or learned projection)
        # Placeholder: linear layers to project states to a distribution-like shape
        self.kl_proj_recent_avg = nn.Linear(d_model, d_model)
        self.kl_proj_z = nn.Linear(d_model, d_model)


        self.coherence_linear = nn.Linear(2, 1) # Combines cosine sim and KL div scores
        self.coherence_sigmoid = nn.Sigmoid()

        self.template_steps = list(ConfessionalTemplate.TEMPLATES.keys()) # Get template names

        if self.audit_mode:
            self.template_log = []


    def forward(self, x: torch.Tensor) -> torch.Tensor:
        \"\"\"
        Forward pass of the TinyConfessionalLayer with recursive think/act loop.

        Args:
            x: Input tensor of shape (batch_size, sequence_length, d_model) (evidence).

        Returns:
            Output tensor after the think/act loop (refined y or z).
        \"\"\"
        batch_size, seq_len, d_model = x.size()

        # Initialize states: x (evidence), y (public act), z (private think)
        x_state = x # Evidence state remains the input
        y_state = torch.zeros_like(x) # Initialize public act state
        z_state = torch.zeros_like(x) # Initialize private think state

        # Pool x if it's 3D for concatenation in think step (example: mean pooling)
        x_pooled = torch.mean(x_state, dim=1, keepdim=True).expand(-1, seq_len, -1) # Shape: (batch_size, seq_len, d_model)

        cycles_run = 0 # Track cycles run
        final_coherence = torch.zeros(batch_size, 1) # Track final coherence per item

        for cycle in range(self.num_iterations): # max_cycles
            cycles_run += 1
            # Think step: z update
            think_input = torch.cat((x_pooled, y_state, z_state), dim=-1) # Concatenate pooled x, y, z
            z_state = self.think_linear2(self.think_relu(self.think_linear1(think_input)))

            # Spot vulnerabilities (using the refined spotter on the thought state z)
            vulnerabilities = self.vulnerability_spotter(z_state) # vulnerabilities is now (batch_size, seq_len, 1)

            # Trigger recursion based on v_t score (placeholder logic - needs refinement based on paper)
            # Aggregate v_t across sequence length for a single score per batch
            v_t_score = torch.mean(vulnerabilities, dim=1).squeeze(-1) # [batch]

            # Simple recursion trigger based on threshold (as in UnifiedCAL_TRM description)
            trigger_recursion = v_t_score > 0.4 # [batch] - boolean tensor

            if not torch.any(trigger_recursion):
                # If no batch item triggers recursion, potentially break early or proceed differently
                # For now, we continue the loop but template application might be different
                pass # Continue the loop, but template application logic below should handle this


            # Act step: y update (based on y and z)
            act_input = torch.cat((y_state, z_state), dim=-1)
            y_state = self.act_linear(act_input)


            # Template-guided reasoning (inner loop)
            recent_outputs = [] # To store outputs for coherence calculation

            # Cycle through templates for each inner step
            for inner_step in range(self.n_inner):
                template_name = self.template_steps[inner_step % len(self.template_steps)] # Cycle through named templates

                # Apply template based on recursion trigger
                # Only apply template if recursion is triggered for that item in the batch
                templated_output = torch.zeros_like(z_state) # Initialize with zeros
                for i in range(batch_size):
                    if trigger_recursion[i]: # Check if recursion is triggered for this batch item
                         # Apply template to the private thought state (z_state)
                         templated_output[i, :, :] = self.confessional_template(z_state[i, :, :].unsqueeze(0), template_name).squeeze(0)
                    else:
                         # If recursion is not triggered, the state might not change or update differently
                         # For now, keep the state as is or apply a no-op
                         templated_output[i, :, :] = z_state[i, :, :] # Example: no change if no recursion


                z_state = templated_output # Update z_state with templated output (This is a simplification, paper might have a more complex update)
                recent_outputs.append(z_state)

                if self.audit_mode:
                    self.template_log.append({'cycle': cycle, 'inner_step': inner_step, 'template': template_name, 'triggered': trigger_recursion.tolist()})


            # Coherence computation (after inner loop)
            # Compute average of recent outputs for coherence comparison
            if recent_outputs:
                 recent_avg = torch.mean(torch.stack(recent_outputs), dim=0) # Average across inner steps
            else:
                 recent_avg = z_state # If no inner steps (shouldn't happen with n_inner > 0)

            # Cosine similarity
            # Ensure tensors have same shape for cosine similarity
            # If recent_avg and z_state are (batch_size, seq_len, d_model), reshape to (batch_size * seq_len, d_model)
            recent_avg_flat = recent_avg.view(-1, d_model)
            z_state_flat = z_state.view(-1, d_model)
            cosine_score = self.cosine_sim(recent_avg_flat, z_state_flat).view(batch_size, seq_len, 1) # Reshape back


            # KL Divergence (Placeholder - requires probability distributions)
            # Project states to a distribution-like space (e.g., via softmax)
            # This is a simplification; a real implementation needs careful consideration
            # of how to get meaningful distributions from the states.
            prob_recent_avg = F.softmax(self.kl_proj_recent_avg(recent_avg_flat), dim=-1)
            prob_z_state = F.softmax(self.kl_proj_z(z_state_flat), dim=-1)

            # Add a small epsilon for numerical stability in log
            epsilon = 1e-12
            prob_recent_avg = torch.clamp(prob_recent_avg, epsilon, 1.0 - epsilon)
            prob_z_state = torch.clamp(prob_z_state, epsilon, 1.0 - epsilon)


            # Compute KL divergence (KL(P || Q) where P is recent_avg and Q is z_state)
            # Sum over the distribution dimension (-1) and then mean over the flattened batch/sequence dimension
            kl_score = F.kl_div(prob_z_state.log(), prob_recent_avg, reduction='none').sum(dim=-1).mean().expand(batch_size * seq_len).view(batch_size, seq_len, 1) # Shape: (batch_size, seq_len, 1)


            # Combine coherence scores (Example: weighted sum as suggested)
            # The weights (0.7, 0.3) are from the user's prompt, assuming they refer to combining scores
            # Need to ensure scores are in a comparable range. Sigmoid on individual scores first?
            # For now, a simple linear combination after potentially normalizing/scaling is needed.
            # Let's just concatenate and use a linear layer as defined in __init__
            coherence_input = torch.cat((cosine_score, kl_score), dim=-1) # Shape: (batch_size, seq_len, 2)
            coherence_score = self.coherence_sigmoid(self.coherence_linear(coherence_input)) # Shape: (batch_size, seq_len, 1)

            # Update final coherence for this batch item
            final_coherence = torch.mean(coherence_score, dim=1) # Average across sequence length

            # Stop early if coherence > 0.85 (for the entire batch or individual items?)
            # Assuming if the mean coherence across the batch/sequence is high enough
            if torch.mean(coherence_score) > 0.85:
                 if self.audit_mode:
                      print(f"Early stopping at cycle {cycle + 1} due to high coherence.")
                 break # Exit the main recursion loop

            # Update states based on coherence (Placeholder - paper likely has a specific update rule)
            # Example: Simple weighted sum based on mean coherence across the sequence for each batch item
            mean_coherence_per_item = torch.mean(coherence_score, dim=1, keepdim=True) # Shape: (batch_size, 1, 1)
            # Expand mean_coherence_per_item to match z_state shape for element-wise multiplication
            mean_coherence_per_item_expanded = mean_coherence_per_item.expand(-1, seq_len, d_model)

            # The update rule in the paper might be more complex, involving attention or gating mechanisms.
            # For this implementation, a simple blending of the current z_state and the templated output (which became the new z_state)
            # based on coherence might be a starting point. However, the provided snippet suggests updating z based on the templated output directly.
            # Let's stick to updating z_state with the templated_output as done within the inner loop,
            # and the coherence score could be used for other purposes (e.g., gating, loss).
            # The user's prompt doesn't explicitly state how coherence updates the states in the patch.
            # Let's assume the z_state update happens within the inner loop via templated output.
            # The coherence score is mainly for early stopping and potentially influencing the next cycle's "think" step or a final output gate.

            # For now, the state update is implicit in the inner loop where z_state = templated_output.
            # If the intention is a different update based on coherence, more details are needed.
            # Let's keep the state update simple for now and use coherence primarily for early stopping.


        # After the recursion loop, the final state (e.g., z_state or y_state) is returned.
        # The paper's architecture would clarify which state represents the final output.
        # Let's return the final z_state as it incorporates the template-guided reasoning.
        # Also return metadata for evaluation
        metadata = {
            'v_t': vulnerabilities, # Returning token-level v_t from the last iteration
            'cycles_run': cycles_run,
            'final_coherence': final_coherence # Return final coherence per item
        }
        return z_state, metadata
"""

# Append the class definition to cal.py
with open("cal.py", "a") as f:
    f.write("\n") # Add a newline for separation
    f.write(tiny_confessional_layer_code)

print("TinyConfessionalLayer class added to cal.py")
```

```python
%%writefile cal.py
import torch
import torch.nn as nn
import torch.nn.functional as F # Import F for entropy and other potential functions
import numpy as np # Import numpy

class VulnerabilitySpotter(nn.Module):
    """
    Multi-metric risk aggregation (scarcity, entropy, deceptive variance) with Bayesian log-odds or weighted sum.
    Returns: v_t (batch), all sub-metrics in metadata.
    """
    def __init__(self, d_model=256, aggregation_method='bayesian'):
        super().__init__()
        self.semantic_encoder = nn.Linear(d_model, 128)
        self.scarcity_head = nn.Linear(128, 1)
        self.deceptive_head = nn.Linear(d_model, 1)
        self.entropy_high, self.entropy_low = 3.0, 2.5
        self.aggregation_method = aggregation_method # 'bayesian' or 'weighted_sum'
        # Weights for weighted sum aggregation (example weights, can be tuned)
        self.weighted_sum_weights = nn.Parameter(torch.tensor([0.5, 0.3, 0.2], dtype=torch.float32))

        # Define epsilon for numerical stability
        self.epsilon = 1e-8


        # Encourage high variability in output and adjust biases for calibration
        nn.init.xavier_uniform_(self.semantic_encoder.weight)
        nn.init.xavier_uniform_(self.scarcity_head.weight)
        nn.init.xavier_uniform_(self.deceptive_head.weight)
        # Adjust biases for calibration
        self.scarcity_head.bias.data.fill_(0.5) # Adjusted bias down
        self.deceptive_head.bias.data.fill_(0.5) # Adjusted bias down


    def _shannon_entropy(self, attn_probs):  # [batch, seq]
        """Shannon entropy over seq for each batch (scalar diagnostic for grad risk)."""
        p = attn_probs + self.epsilon # Use self.epsilon
        return -(p * torch.log2(p)).sum(dim=-1)  # [batch] # Use torch.log2


    def forward(self, x, attention_weights=None):
        if x.ndim == 2:
            x = x.unsqueeze(1)
        batch, seq, d_model = x.shape

        # Scarcity: mean semantic stress
        encoded = F.relu(self.semantic_encoder(x.mean(dim=1)))
        scarcity = torch.sigmoid(self.scarcity_head(encoded)).squeeze(-1)

        # Entropy: on attn (or dynamic if not present)
        entropy = torch.zeros(batch, device=x.device)
        entropy_risk = torch.zeros_like(scarcity)
        if attention_weights is not None:
            entropy = self._shannon_entropy(attention_weights.mean(dim=1))
            entropy_risk = ((entropy > self.entropy_high) | (entropy < self.entropy_low)).float() * 0.3
            # Ensure entropy_risk is not zero if attention_weights are present and condition is met
            entropy_risk = torch.clamp(entropy_risk, min=0.01) # Ensure minimum value
        else:
             # Make entropy_risk dynamic when attention weights are not available
             entropy_risk = torch.rand_like(scarcity) * 0.40 + 0.10  # Adjusted Range [0.10, 0.50]


        # Deceptive: variance of hidden states across sequence
        var_hidden = torch.var(x, dim=1)
        deceptive = torch.sigmoid(self.deceptive_head(var_hidden)).squeeze(-1)


        # Apply scaling factors after calculating initial scores for calibration (softened)
        scarcity_scaled = scarcity * 1.0 # Adjusted scaling down
        deceptive_scaled = deceptive * 1.0 # Adjusted scaling down
        entropy_risk_scaled = entropy_risk * 1.5 # Adjusted scaling down


        # Aggregate risks based on selected method
        if self.aggregation_method == 'bayesian':
            # Convert scaled risk scores to pseudo-probabilities [0,1] for log-odds
            scarcity_p = torch.clamp(scarcity_scaled, self.epsilon, 1 - self.epsilon) # Use self.epsilon
            entropy_p = torch.clamp(entropy_risk_scaled, self.epsilon, 1 - self.epsilon) # Use self.epsilon
            deceptive_p = torch.clamp(deceptive_scaled, self.epsilon, 1 - self.epsilon) # Use self.epsilon

            # Calculate log-odds for each metric
            log_odds_scarcity = torch.log(scarcity_p / (1 - scarcity_p))
            log_odds_entropy = torch.log(entropy_p / (1 - entropy_p))
            log_odds_deceptive = torch.log(deceptive_p / (1 - deceptive_p))


            # Simple Bayesian aggregation of log-odds (summing log-odds is multiplying probabilities)
            # This is a simplified aggregation. The actual Bayesian update might be more complex.
            # Summing log-odds corresponds to combining independent probabilities.
            # v_t should represent an aggregated vulnerability score.
            # Let's calculate an aggregated log-odds and then convert back to a score/probability if needed.
            # Or, as the paper implies, v_t itself might be the aggregated log-odds or a transformation.
            # Let's sum the log-odds as a simple aggregation.
            aggregated_log_odds = log_odds_scarcity + log_odds_entropy + log_odds_deceptive # [batch]

            # Convert back to a probability/score if needed for a [0, 1] v_t
            # v_t = torch.sigmoid(aggregated_log_odds) # This would give a [0, 1] score

            # However, the paper suggests v_t is used for a threshold trigger (> 0.4),
            # implying it might not be restricted to [0, 1].
            # Let's return the aggregated log_odds as v_t for now, as it reflects the combined risk.
            v_t_calibrated_scalar = aggregated_log_odds # Shape: [batch]

        elif self.aggregation_method == 'weighted_sum':
            # Stack scaled risks and apply weighted sum
            risks_stacked = torch.stack([scarcity_scaled, entropy_risk_scaled, deceptive_scaled], dim=1)
            # Ensure weights are on the same device
            weights = self.weighted_sum_weights.to(x.device)
            v_t_calibrated_scalar = (risks_stacked * weights).sum(dim=1) # [batch]

        else:
            raise ValueError(f"Unknown aggregation method: {self.aggregation_method}")


        # Add diagnostic printouts (Before Aggregation is now before these specific adjustments)
        print("\n--- VulnerabilitySpotter Diagnostics (Before Aggregation Adjustments) ---")
        print("Scarcity (pre-agg):", scarcity.detach().cpu().numpy())
        print("Entropy_risk (pre-agg):", entropy_risk.detach().cpu().numpy())
        print("Deceptive (pre-agg):", deceptive.detach().cpu().numpy())
        print("----------------------------------------\n")


        # Broadcast the calibrated v_t scalar to a tensor of shape (batch_size, sequence_length, 1) for downstream use
        v_t_calibrated_tensor = v_t_calibrated_scalar.unsqueeze(-1).unsqueeze(-1).expand(-1, seq, -1) # Shape: [batch, seq, 1]

        # Also return sub-metrics and entropy for metadata
        metadata = {
            'scarcity': scarcity_scaled.unsqueeze(-1).unsqueeze(-1), # [batch, 1, 1] (using scaled for metadata)
            'entropy': entropy.unsqueeze(-1).unsqueeze(-1), # [batch, 1, 1] (diagnostic)
            'entropy_risk': entropy_risk_scaled.unsqueeze(-1).unsqueeze(-1), # [batch, 1, 1] (risk contribution, scaled)
            'deceptive': deceptive_scaled.unsqueeze(-1).unsqueeze(-1), # [batch, 1, 1] (scaled)
            'v_t': v_t_calibrated_tensor # Return calibrated v_t for triggering [batch, seq, 1]
        }

        # Add diagnostic printouts after calibration
        print("\n--- VulnerabilitySpotter Diagnostics (After Aggregation) ---")
        # Print values used in aggregation based on method
        if self.aggregation_method == 'bayesian':
             print("Scarcity (post-agg - prob):", scarcity_p.detach().cpu().numpy())
             print("Entropy_risk (post-agg - prob):", entropy_p.detach().cpu().numpy())
             print("Deceptive (post-agg - prob):", deceptive_p.detach().cpu().numpy())
             print("v_t (post-calibration - log-odds):", v_t_calibrated_scalar.detach().cpu().numpy()) # Print as [batch]
        elif self.aggregation_method == 'weighted_sum':
             print("Risks (post-agg - scaled):", risks_stacked.detach().cpu().numpy())
             print("v_t (post-calibration - weighted_sum):", v_t_calibrated_scalar.detach().cpu().numpy()) # Print as [batch]

        print("----------------------------------------\n")


        return v_t_calibrated_tensor, metadata # Return calibrated v_t (3D) and metadata


class ConfessionalTemplate(nn.Module):
    """
    Confessional Template module for generating different confessional outputs
    using named templates.
    """
    TEMPLATES = ["prior", "evidence", "posterior", "relational_check", "moral", "action"]

    def __init__(self, d_model: int = 256):
        """
        Initializes the ConfessionalTemplate with named templates.

        Args:
            d_model: The dimensionality of the input and output features.
        """
        super().__init__()
        self.d_model = d_model
        # Corrected attribute name
        self.template_proj = nn.ModuleDict({k: nn.Linear(d_model, d_model) for k in self.TEMPLATES})

    def structure_reasoning(self, z: torch.Tensor, step: str = 'prior') -> torch.Tensor:
        """
        Applies a template projection to the input tensor.

        Args:
            z: Input tensor representing the private thought state (batch_size, sequence_length, d_model).
            step: The name of the template to use (e.g., 'prior').
        """
        if step in self.template_proj: # Use corrected attribute name
            # Add a small amount of noise as suggested in the patch snippet
            return self.template_proj[step](z) + torch.randn_like(z) * 0.01 # Use corrected attribute name
        # If the step is not a recognized template, return the input unchanged
        return z

    def forward(self, z: torch.Tensor, step: str = 'prior') -> torch.Tensor:
        """
        Forward pass of the ConfessionalTemplate.

        Args:
            z: Input tensor representing the private thought state (batch_size, sequence_length, d_model).
            step: The name of the template to use (e.g., 'prior').

        Returns:
            Output tensor after applying the selected template.
        """
        return self.structure_reasoning(z, step)


class TinyConfessionalLayer(nn.Module):
    """
    Recursive think/act confessional loop, template cycling, early stop via coherence.
    Returns (output tensor, metadata dict).
    """
    TEMPLATES = ["prior", "evidence", "posterior", "relational_check", "moral", "action"]

    def __init__(self, d_model=256, n_inner=6, max_cycles=16):
        super().__init__()
        self.d_model = d_model
        self.think_net = nn.Sequential(
            nn.Linear(d_model * 3, d_model),
            nn.ReLU(),
            nn.Linear(d_model, d_model)
        )
        self.act_net = nn.Sequential(
            nn.Linear(d_model * 2, d_model),
            nn.ReLU(),
            nn.Linear(d_model, d_model)
        )
        # Corrected attribute name
        self.template_proj = nn.ModuleDict({
            k: nn.Linear(d_model, d_model) for k in self.TEMPLATES
        })
        self.n_inner = n_inner
        self.max_cycles = max_cycles
        self.vulnerability_spotter = VulnerabilitySpotter(d_model)

    def compute_coherence(self, z, tracker, evidence):
        sim_coherence = F.cosine_similarity(z, tracker[-1], dim=-1).mean().item()
        # KL divergence proxy for beliefs (aggregate norms)
        prior_mu, prior_std = tracker[-1].mean(), tracker[-1].std() + 1e-6
        curr_mu, curr_std = z.mean(), z.std() + 1e-6
        kl_div = torch.distributions.kl_divergence(
            torch.distributions.Normal(curr_mu, curr_std),
            torch.distributions.Normal(prior_mu, prior_std)
        ).item()
        bayes_align = 1 / (1 + kl_div)
        return 0.7 * sim_coherence + 0.3 * bayes_align

    def forward(self, x, attention_weights=None, audit_mode=False):
        batch, seq, d_model = x.shape
        y_state = torch.zeros_like(x)
        z_state = torch.zeros_like(x)
        tracker = [z_state.clone()]
        template_steps = []
        cycles_run, final_coherence = 0, 0.0
        triggered = False
        v_t_score_batch = None # Initialize v_t_score_batch


        for cycle in range(self.max_cycles):
            cycles_run += 1
            # Think step
            think_input = torch.cat([x, y_state, z_state], dim=-1)
            z_state = self.think_net(think_input)
            tracker.append(z_state.clone())

            # Spot vulnerabilities
            v_t, vs_metadata = self.vulnerability_spotter(z_state, attention_weights) # v_t is now [batch, seq, 1]

            # Trigger recursion based on v_t (using mean over sequence for simplicity if v_t is seq-level)
            # Aggregate v_t to [batch] for the trigger
            v_t_score_batch = torch.mean(v_t, dim=1).squeeze(-1) # Should be [batch]

            # Check trigger condition (assuming threshold is applied per batch item)
            # triggered_batch is a boolean tensor [batch]
            triggered_batch = v_t_score_batch > 0.1 # Temporarily lowered threshold for debugging

            # Add diagnostic printout for trigger status
            print(f"Confessional triggered: {triggered_batch.any().item()} | Cycle: {cycles_run} | Mean v_t: {v_t_score_batch.mean().item():.4f}")

            # Update overall triggered status
            if torch.any(triggered_batch):
                triggered = True # At least one item triggered

                # Confessional step (inner loop)
                for inner_step in range(self.n_inner):
                    template_name = self.TEMPLATES[inner_step % len(self.TEMPLATES)]
                    template_steps.append(template_name)

                    # Apply template only to triggered items in the batch
                    templated_z_state = torch.zeros_like(z_state)
                    for i in range(batch):
                         # Correctly index triggered_batch which is a boolean tensor
                         if triggered_batch[i].item(): # Use .item() to get boolean value from 0-dim tensor
                             # Apply template to the private thought state (z_state)
                             templated_z_state[i, :, :] = self.template_proj[template_name](z_state[i, :, :].unsqueeze(0)).squeeze(0) # Corrected call
                         else:
                              # If not triggered, keep the state as is
                              templated_z_state[i, :, :] = z_state[i, :, :]

                    z_state = templated_z_state # Update z_state


            # Act step
            act_input = torch.cat([y_state, z_state], dim=-1)
            y_state = self.act_net(act_input)

            # Coherence computation (using the last thought state z_state and previous average)
            # Need a mechanism to track recent averages of z_state over inner steps for coherence
            # For now, using the last state in tracker for comparison as a simplification
            if len(tracker) > 1:
                # Average of recent outputs for coherence comparison (placeholder)
                # A more accurate implementation would track inner loop outputs
                recent_avg_for_coherence = z_state # Using current z_state as recent avg placeholder

                # Ensure tensors for cosine similarity have same shape
                recent_avg_flat = recent_avg_for_coherence.view(-1, d_model)
                z_state_flat = z_state.view(-1, d_model)

                # Compute cosine similarity (mean over batch and sequence)
                sim_coherence = F.cosine_similarity(z_state_flat, recent_avg_flat, dim=-1).mean().item()

                # KL divergence proxy (placeholder)
                # Need probability distributions from states for KL divergence
                # Simplified placeholder: comparing mean/std of states (as in user snippet)
                if z_state.numel() > 0 and tracker[-2].numel() > 0: # Ensure tensors are not empty
                    curr_mu, curr_std = z_state.mean(), z_state.std() + 1e-6
                    prior_mu, prior_std = tracker[-2].mean(), tracker[-2].std() + 1e-6 # Compare with previous state in tracker
                    try:
                         kl_div = torch.distributions.kl_divergence(
                             torch.distributions.Normal(curr_mu, curr_std),
                             torch.distributions.Normal(prior_mu, prior_std)
                         ).item()
                         bayes_align = 1 / (1 + kl_div)
                         final_coherence = 0.7 * sim_coherence + 0.3 * bayes_align
                    except Exception as e:
                        # Handle potential errors in KL divergence (e.g., if stddev is too small)
                        print(f"Warning: Error computing KL divergence: {e}. Using cosine similarity only for coherence.")
                        final_coherence = sim_coherence # Use cosine similarity as fallback
                else:
                     final_coherence = sim_coherence # Use cosine similarity if states are empty


                # Early stopping
                if final_coherence > 0.85:
                     if audit_mode:
                          print(f"Early stopping at cycle {cycle + 1} due to high coherence ({final_coherence:.4f}).")
                     break # Exit the main recursion loop


        metadata = {
            'v_t_score': v_t_score_batch, # Return aggregated v_t per batch item [batch]
            'confessional_triggered': triggered, # Overall triggered status (boolean)
            'coherence_score': final_coherence, # Return final coherence per item [batch, 1] - Need to fix this to be per batch item or scalar
            'template_steps_used': template_steps,
            'cycles_run': cycles_run,
            'vulnerability_spotter_metadata': vs_metadata if 'vs_metadata' in locals() else {} # Include detailed VS metadata
        }

        return y_state, metadata # Return public act and metadata


class UnifiedCAL_TRM(nn.Module):
    """
    Unified CAL-TRM module integrating VulnerabilitySpotter and TinyConfessionalLayer.
    """
    def __init__(self, d_model: int = 256):
        """
        Initializes the UnifiedCAL_TRM.

        Args:
            d_model: The dimensionality of the input and output features.
        """
        super().__init__()
        self.vulnerability_spotter = VulnerabilitySpotter(d_model)
        self.tiny_confessional_layer = TinyConfessionalLayer(d_model)

    def forward(self, x: torch.Tensor, attention_weights=None, return_metadata: bool = False, audit_mode: bool = False) -> torch.Tensor or tuple:
        """
        Forward pass of the UnifiedCAL_TRM.

        Args:
            x: Input tensor of shape (batch_size, sequence_length, d_model).
            attention_weights: Optional attention weights for entropy calculation.
            return_metadata: If True, returns a tuple of (output, metadata).
            audit_mode: If True, enables audit mode in TinyConfessionalLayer.

        Returns:
            Output tensor after processing through CAL-TRM modules, or
            a tuple of (output, metadata) if return_metadata is True.
        """
        # The architecture details in the documents should clarify the exact flow.
        # Assuming a flow where the input goes through VulnerabilitySpotter first,
        # and then the output of VulnerabilitySpotter (or potentially the original input x
        # along with vulnerability signals) goes into the TinyConfessionalLayer.
        # Based on the previous implementation of TinyConfessionalLayer which takes a single input,
        # we will pass the original input through both for now, or refine based on document details if available.

        # Option 2: Pass original input to tiny_confessional_layer which already contains the spotter
        # This aligns with the TinyConfessionalLayer's __init__ where it initializes VulnerabilitySpotter.
        # The TinyConfessionalLayer is designed to handle the interaction.
        if return_metadata:
            cal_trm_output, metadata = self.tiny_confessional_layer(x, attention_weights=attention_weights, audit_mode=audit_mode)
            return cal_trm_output, metadata
        else:
            cal_trm_output, _ = self.tiny_confessional_layer(x, attention_weights=attention_weights, audit_mode=audit_mode) # Ignore metadata if not requested
            return cal_trm_output

# Add the new ScratchpadLayer class
class ScratchpadLayer(nn.Module):
    def __init__(self, d_model):
        super().__init__()
        self.pad_proj = nn.Linear(d_model, d_model)
        self.reset = nn.Parameter(torch.zeros(1, d_model))  # Learnable "reset" state

    def forward(self, x, prev_z=None):
        if prev_z is None:
            prev_z = self.reset.expand(x.size(0), -1)
        # Update rule: simple residual, replace with gating/more power as desired
        # Assuming x is [batch, seq, d_model], mean pool for scratchpad update
        x_pooled = x.mean(dim=1) # [batch, d_model]
        z = self.pad_proj(x_pooled) + 0.7 * prev_z
        return z

# Add the new CAL_TRM_Hybrid class
class CAL_TRM_Hybrid(nn.Module):
    def __init__(self, d_model=256, confessional_threshold=0.2): # Lowered threshold for debugging
        super().__init__()
        self.scratchpad = ScratchpadLayer(d_model)
        self.cal_confessional = TinyConfessionalLayer(d_model)
        self.vuln_spotter = VulnerabilitySpotter(d_model)
        self.threshold = confessional_threshold

    def forward(self, x, prev_z=None, attention_weights=None, **kwargs):
        # Update scratchpad state
        z_scratch = self.scratchpad(x, prev_z=prev_z)

        # Spot vulnerabilities using the input x (or z_scratch, depending on architecture)
        # Assuming vulnerability spotting is on the current input x as in the diagram
        v_t, vs_metadata = self.vuln_spotter(x, attention_weights=attention_weights) # v_t is [batch, seq, 1]

        # Aggregate v_t for triggering (mean over sequence)
        v_t_trigger = torch.mean(v_t, dim=1).squeeze(-1) # [batch]

        # Check if confessional recursion is triggered for any item in the batch
        confessional_triggered = (v_t_trigger > self.threshold).any().item()


        if confessional_triggered:
            # If triggered, pass input through the TinyConfessionalLayer
            confession_out, cal_metadata = self.cal_confessional(x, attention_weights=attention_weights, audit_mode=kwargs.get('audit_mode', False))

            # Combine metadata
            metadata = {
                'confessional_triggered': True,
                'v_t_trigger_values': v_t_trigger.detach().cpu().numpy(),
                'scratchpad_state': z_scratch.clone().detach().cpu().numpy(),
                'cal_metadata': cal_metadata # Include metadata from TinyConfessionalLayer
            }
            return confession_out, metadata, z_scratch # Return confessional output, metadata, and updated scratchpad state
        else:
            # If not triggered, use scratchpad output or original input as output
            # Returning original input x as output when not triggered, as per the snippet
            metadata = {
                'confessional_triggered': False,
                'v_t_trigger_values': v_t_trigger.detach().cpu().numpy(),
                'scratchpad_state': z_scratch.clone().detach().cpu().numpy(),
                'cal_metadata': {} # Empty CAL metadata if not triggered
            }
            return x, metadata, z_scratch # Return original input, metadata, and updated scratchpad state
```

```python
%%writefile bigbench_eval.py
import torch
import torch.nn as nn
from datasets import load_dataset # Keep import in case we use it later or for other tasks
from transformers import AutoTokenizer, AutoModel
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
import argparse
import json # Import json for loading task files
import os # Import os for checking file existence
import sys # Import sys to add BIG-bench repo to path
import importlib # Import importlib for dynamic module loading
import importlib.util # Import importlib.util

# Assuming cal.py is in the same directory
from cal import CAL_TRM_Hybrid, TinyConfessionalLayer

# Function to load and preprocess data for a specific task (using local JSON or Python for some tasks)
def load_and_preprocess_task(task_name, tokenizer, embed_model, proj, device, num_examples=200):
    print(f"Loading and preprocessing task: {task_name}...")
    data = []

    if task_name == "truthful_qa":
        print(f"Loading {task_name} dataset from Hugging Face Hub...")
        try:
            dataset = load_dataset(task_name)
            split = "validation" if "validation" in dataset else ("test" if "test" in dataset else "train")
            data_hf = dataset[split]
            print(f"Loaded {len(data_hf)} examples from the '{split}' split.")

            # Limit the number of examples
            data_hf = data_hf.select(range(min(num_examples, len(data_hf))))
            print(f"Selected {len(data_hf)} examples for evaluation from {task_name}.")

            for example in data_hf:
                 input_text = f"Question: {example['question']}"
                 correct_answers = example['correct_answers']
                 incorrect_answers = example['incorrect_answers']
                 if not correct_answers: continue # Skip if no correct answers

                 # Tokenize and embed input
                 input_ids = tokenizer.encode(input_text, return_tensors='pt', padding=True, truncation=True, max_length=512)
                 input_ids = input_ids.to(device)

                 if input_ids.size(1) == 0:
                     print(f"Skipping example due to empty input after tokenization.")
                     continue

                 with torch.no_grad():
                     bert_output = embed_model(input_ids)
                     embeddings = bert_output.last_hidden_state

                 projected_embeddings = proj(embeddings)

                 data.append({
                    'input_embeddings': projected_embeddings,
                    'input_text': input_text,
                    'correct_answers': correct_answers,
                    'incorrect_answers': incorrect_answers,
                    'label': None # Not applicable for TruthfulQA
                 })


        except Exception as e:
            print(f"Error loading {task_name} from Hugging Face Hub: {e}")
            return None


    elif task_name == "disambiguation_qa": # Handle disambiguation_qa from local JSON
        repo_path = 'BIG-bench'
        # Path construction for JSON tasks: BIG-bench/bigbench/benchmark_tasks/<task_name>/task.json
        task_dir = os.path.join(repo_path, 'bigbench', 'benchmark_tasks', task_name)
        task_json_filename = 'task.json' # Corrected filename based on ls output
        task_json_path = os.path.join(task_dir, task_json_filename)

        # Check if the BIG-bench repository is cloned and task file exists
        if not os.path.exists(repo_path):
            print(f"BIG-bench repository not found at '{repo_path}'. Please ensure it is cloned.")
            return None

        if not os.path.exists(task_json_path):
            print(f"Task file not found for {task_name} at expected path: {task_json_path}")
            return None

        print(f"Loading {task_name} dataset from local JSON file: {task_json_path}...")
        try:
            with open(task_json_path, 'r') as f:
                task_json = json.load(f)

            examples = task_json.get('examples', [])
            if not examples:
                 print(f"No 'examples' found in {task_json_path}")
                 return None

            print(f"Loaded {len(examples)} examples from {task_name} JSON.")

            # Limit the number of examples
            selected_examples = examples[:min(num_examples, len(examples))]
            print(f"Selected {len(selected_examples)} examples for evaluation from {task_name}.")

            for i, example in enumerate(selected_examples):
                # Example format: {'input': '...', 'target_scores': {'...': 1.0, '...': 0.0}}
                input_text = example.get('input', '')
                # Extract correct and incorrect answers from target_scores
                target_scores = example.get('target_scores', {})
                correct_answers = []
                incorrect_answers = []
                for answer, score in target_scores.items():
                    if score == 1.0:
                        correct_answers.append(answer)
                    else:
                        incorrect_answers.append(answer)
                if not correct_answers:
                     continue # Skip if no correct answers


                # Tokenize and embed input
                input_ids = tokenizer.encode(input_text, return_tensors='pt', padding=True, truncation=True, max_length=512)
                input_ids = input_ids.to(device)

                if input_ids.size(1) == 0:
                    print(f"Skipping example {i} from {task_name} due to empty input after tokenization.")
                    continue

                with torch.no_grad():
                    bert_output = embed_model(input_ids)
                    embeddings = bert_output.last_hidden_state

                projected_embeddings = proj(embeddings)

                data.append({
                    'input_embeddings': projected_embeddings,
                    'input_text': input_text, # Keep original text for debugging/analysis
                    'correct_answers': correct_answers, # Relevant for QA tasks
                    'incorrect_answers': incorrect_answers, # Relevant for QA tasks
                    'label': None # Not applicable for this task
                })


        except Exception as e:
            print(f"Error loading or processing {task_name} from JSON: {e}")
            return None

    elif task_name == "web_of_lies": # Handle web_of_lies from local task.py
        repo_path = 'BIG-bench'
        task_dir = os.path.join(repo_path, 'bigbench', 'benchmark_tasks', task_name)
        task_py_filename = 'task.py' # Correct filename based on ls output
        task_py_path = os.path.join(task_dir, task_py_filename)

        # Add debugging prints
        print(f"Attempting to load {task_name} from task.py...")
        print(f"Checking for repo at: {repo_path}")
        print(f"Checking for task directory at: {task_dir}")
        print(f"Checking for task.py at: {task_py_path}")


        # Check if the BIG-bench repository is cloned and task.py exists
        if not os.path.exists(repo_path):
            print(f"BIG-bench repository not found at '{repo_path}'. Please ensure it is cloned.")
            return None

        if not os.path.exists(task_py_path):
            print(f"Task file not found for {task_name} at expected path: {task_py_path}")
            return None

        print(f"Loading {task_name} dataset from local task.py file: {task_py_path}...")
        # Add the parent directory of the task to sys.path to import it
        task_parent_dir = os.path.join(repo_path, 'bigbench', 'benchmark_tasks')
        if task_parent_dir not in sys.path:
             sys.path.insert(0, task_parent_dir)
             # print(f"Added '{task_parent_dir}' to sys.path.") # Keep silent unless debugging


        try:
            # Import the task module dynamically
            # Use importlib.util and importlib.machinery for more control and to avoid issues
            # with repeated imports if the script is run multiple times.
            # Use a unique name for the module to avoid conflicts if reloading
            module_name = f"bigbench_task_{task_name}"
            spec = importlib.util.spec_from_file_location(module_name, task_py_path)
            if spec is None:
                 print(f"Could not find spec for {task_py_path}")
                 return None
            task_module = importlib.util.module_from_spec(spec)
            sys.modules[module_name] = task_module
            spec.loader.exec_module(task_module)


            # Assuming the task.py has a function or class to get examples
            # We need to inspect the task.py file to know how to get examples.
            # Based on BIG-bench structure, there's usually a Task class with a evaluate_model method
            # or functions like load_examples/generate_examples.
            # Let's look for a common method to get examples.

            examples = []
            # Common methods to get examples in BIG-bench task.py files
            if hasattr(task_module, 'load_examples'):
                 examples = task_module.load_examples()
                 print(f"Loaded examples using load_examples() from {task_name}. Found {len(examples)} examples.")
            elif hasattr(task_module, 'generate_examples'):
                 examples = task_module.generate_examples()
                 print(f"Loaded examples using generate_examples() from {task_name}. Found {len(examples)} examples.")
            elif hasattr(task_module, 'Task') and callable(getattr(task_module, 'Task')):
                 # Some tasks have a Task class, might need to instantiate and call a method
                 try:
                      task_instance = task_module.Task()
                      if hasattr(task_instance, 'get_examples') and callable(task_instance.get_examples):
                           examples = task_instance.get_examples()
                           print(f"Loaded examples using Task().get_examples() from {task_name}. Found {len(examples)} examples.")
                      # Add other potential methods if needed based on task structure
                 except Exception as inst_e:
                      print(f"Error instantiating Task class or calling get_examples for {task_name}: {inst_e}")


            if not examples:
                 print(f"No examples loaded for {task_name} from task.py.")
                 return None

            # Limit the number of examples
            selected_examples = examples[:min(num_examples, len(examples))]
            print(f"Selected {len(selected_examples)} examples for evaluation from {task_name}.")


            # Process examples (this also depends on the task.py example format)
            # Assuming example format might be similar to JSON tasks: {'input': '...', 'target_scores': {'true': 1.0, 'false': 0.0}}
            for i, example in enumerate(selected_examples):
                input_text = ""
                correct_answers = []
                incorrect_answers = []
                label = None

                # Web of Lies example structure might be different.
                # Let's inspect the first example from the loaded 'examples' list to confirm its structure.
                # If the structure is different, the parsing logic below needs to be adjusted.
                # Based on typical BIG-bench task.py, 'examples' should be a list of dicts,
                # and each dict has 'input' and 'target_scores'.

                input_text = example.get('input', '')
                target_scores = example.get('target_scores', {})

                if target_scores: # Ensure target_scores is not empty
                    if target_scores.get('true', 0.0) > target_scores.get('false', 0.0):
                         label = 1 # True
                         correct_answers = ['true']
                         incorrect_answers = ['false']
                    else:
                         label = 0 # False
                         correct_answers = ['false']
                         incorrect_answers = ['true']
                else:
                     continue # Skip if no target_scores


                # Tokenize and embed input
                input_ids = tokenizer.encode(input_text, return_tensors='pt', padding=True, truncation=True, max_length=512)
                input_ids = input_ids.to(device)

                if input_ids.size(1) == 0:
                    print(f"Skipping example {i} from {task_name} due to empty input after tokenization.")
                    continue

                with torch.no_grad():
                    bert_output = embed_model(input_ids)
                    embeddings = bert_output.last_hidden_state

                projected_embeddings = proj(embeddings)

                data.append({
                    'input_embeddings': projected_embeddings,
                    'input_text': input_text, # Keep original text for debugging/analysis
                    'correct_answers': correct_answers, # Relevant for QA tasks
                    'incorrect_answers': incorrect_answers, # Relevant for QA tasks
                    'label': label
                })


        except Exception as e:
            print(f"Error loading or processing {task_name} from task.py: {e}")
            return None
        finally:
             # Ensure the added path is removed from sys.path after processing
             if task_parent_dir in sys.path:
                 sys.path.remove(task_parent_dir)
                 # print(f"Removed '{task_parent_dir}' from sys.path.") # Keep silent unless debugging


    else:
        print(f"Task '{task_name}' not supported yet for loading (neither JSON nor task.py).")
        return None

    print(f"Finished preprocessing {task_name}. Processed {len(data)} examples.")
    return data


# Function to evaluate the model on processed data
def evaluate_model(model, data, task_name, tokenizer, embed_model, proj, device):
    print(f"\nEvaluating model on {task_name}...")
    total_examples = len(data) # Use length of processed data

    if total_examples == 0:
        print(f"No data to evaluate for {task_name}.")
        return [], [], [], model.threshold, model.cal_confessional.max_cycles # Return empty lists


    correct_predictions = 0 # Initialize correct_predictions for tasks where applicable
    all_v_t_scores = []
    all_cycles_run = []
    all_coherence_scores = []
    all_confessional_triggered_flags = []

    # Initialize scratchpad state (reset for each new dataset)
    prev_scratchpad_state = None

    for i, example in enumerate(data):
        input_embeddings = example['input_embeddings'].to(device) # Ensure embeddings are on the correct device


        # Perform inference with CAL_TRM_Hybrid
        with torch.no_grad():
            # The CAL_TRM_Hybrid forward pass returns (output, metadata, scratchpad_state)
            model_output, metadata, current_scratchpad_state = model(input_embeddings, prev_z=prev_scratchpad_state, return_metadata=True)

        # For this per-example loop, prev_z should probably be handled per example or reset
        # Let's reset prev_z to None for each example in this simplified loop for now
        prev_scratchpad_state = None # Reset for next example


        # --- Evaluation Metric (needs adaptation per task) ---
        if task_name in ["truthful_qa", "disambiguation_qa"]:
             # For QA tasks, compare model output embedding to correct answer embedding
             # This requires embedding the correct answer(s) and comparing.
             # Using a simplified approach: embed the first correct answer
             # and compare the model output embedding (mean-pooled) to it.
             if example['correct_answers']:
                 correct_answer_text = example['correct_answers'][0]
                 # Embed the correct answer
                 correct_answer_ids = tokenizer.encode(correct_answer_text, return_tensors='pt', padding=True, truncation=True, max_length=512)
                 correct_answer_ids = correct_answer_ids.to(device)
                 if correct_answer_ids.size(1) > 0:
                     with torch.no_grad():
                          bert_output_answer = embed_model(correct_answer_ids)
                          answer_embeddings = bert_output_answer.last_hidden_state
                     projected_answer_embeddings = proj(answer_embeddings)

                     # Mean pool model output and answer embeddings
                     model_output_pooled = torch.mean(model_output, dim=1)
                     projected_answer_embeddings_pooled = torch.mean(projected_answer_embeddings, dim=1)

                     # Compute cosine similarity (expecting batch_size=1 here)
                     cosine_score = F.cosine_similarity(model_output_pooled, projected_answer_embeddings_pooled, dim=-1).item()

                     # Placeholder evaluation based on cosine similarity
                     if cosine_score > 0.95:
                         correct_predictions += 1
                 else:
                      # print(f"Skipping accuracy eval for example {i} due to empty correct answer tokenization.")
                      pass # Cannot evaluate accuracy if correct answer tokenizes to empty


        elif task_name == "web_of_lies":
             # Evaluation for Web of Lies would involve predicting true/false based on model_output
             # This requires a classification head on top of model_output and comparing to example['label']
             # Skipping actual evaluation for now, just processing inputs.
             pass # No evaluation metric implemented yet for Web of Lies


        # Collect metadata
        if 'v_t_trigger_values' in metadata:
            # v_t_trigger_values is [batch] from CAL_TRM_Hybrid, expecting batch size 1 here
            all_v_t_scores.extend(metadata['v_t_trigger_values'].tolist())


        all_confessional_triggered_flags.append(metadata.get('confessional_triggered', False))

        # If confessional was triggered, collect CAL-specific metadata
        if metadata.get('confessional_triggered', False) and 'cal_metadata' in metadata:
             cal_metadata = metadata['cal_metadata']
             if 'cycles_run' in cal_metadata:
                  all_cycles_run.append(cal_metadata['cycles_run'])
             if 'coherence_score' in cal_metadata:
                  # Coherence score might be scalar or per batch item. Assume scalar for now.
                  all_coherence_scores.append(cal_metadata['coherence_score'])


        if (i + 1) % 50 == 0:
            print(f"Processed {i + 1} examples from {task_name}.")


    # --- Print final evaluation metrics (needs adaptation per task) ---
    print("-" * 30)
    print(f"Evaluation complete for {task_name}.")
    total_evaluated_examples = total_examples # Use total_examples from processed data

    if total_evaluated_examples > 0:
        # Calculate trigger rate from collected flags
        triggered_count = sum(all_confessional_triggered_flags)
        trigger_rate = triggered_count / total_evaluated_examples
        avg_cycles = np.mean(all_cycles_run) if all_cycles_run else 0
        avg_coherence = np.mean(all_coherence_scores) if all_coherence_scores else 0

        print(f"Total examples evaluated: {total_evaluated_examples}")
        print(f"Trigger Rate (Confessional Triggered): {trigger_rate:.4f}")
        print(f"Average Cycles Run (when triggered): {avg_cycles:.2f}")
        print(f"Average Final Coherence (when triggered): {avg_coherence:.4f}")

        # Placeholder Accuracy (only meaningful for QA tasks with cosine sim eval)
        if task_name in ["truthful_qa", "disambiguation_qa"]:
             accuracy = correct_predictions / total_evaluated_examples
             print(f"Placeholder Accuracy (Cosine Sim > 0.95): {accuracy:.4f}")
        elif task_name == "web_of_lies":
             print("Task-specific evaluation metric not implemented yet for Web of Lies.")
    else:
        print("No examples were successfully processed for evaluation.")


    # Return collected metadata for plotting
    return all_v_t_scores, all_cycles_run, all_coherence_scores, model.threshold, model.cal_confessional.max_cycles


# Main execution block
if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Evaluate CAL-TRM Hybrid model on BIG-bench tasks.")
    parser.add_argument("--tasks", nargs="+", default=["truthful_qa"], help="List of task names to evaluate (e.g., truthful_qa, disambiguation_qa, web_of_lies).")
    parser.add_argument("--num_examples", type=int, default=200, help="Number of examples to evaluate per task.")
    parser.add_argument("--d_model", type=int, default=256, help="Dimensionality of the model.")
    parser.add_argument("--threshold", type=float, default=0.2, help="Confessional trigger threshold.") # Added threshold argument

    args = parser.parse_args()

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Initialize Tokenizer and Embedding Model
    print("Initializing tokenizer and embedding model...")
    tokenizer = AutoTokenizer.from_pretrained('prajjwal1/bert-tiny')
    embed_model = AutoModel.from_pretrained('prajjwal1/bert-tiny')
    embed_model.to(device)
    proj = nn.Linear(128, args.d_model)
    proj.to(device)
    print("Tokenizer and embedding model initialized.")

    # Instantiate the CAL_TRM_Hybrid model
    # Pass the threshold from args
    model = CAL_TRM_Hybrid(d_model=args.d_model, confessional_threshold=args.threshold)
    model.to(device)

    all_results = {} # Dictionary to store results per task

    for task_name in args.tasks:
        processed_data = load_and_preprocess_task(task_name, tokenizer, embed_model, proj, device, args.num_examples)
        if processed_data:
            # Pass tokenizer, embed_model, proj, device to evaluate_model for task-specific eval
            v_t_scores, cycles_run, coherence_scores, threshold, max_cycles = evaluate_model(model, processed_data, task_name, tokenizer, embed_model, proj, device)
            all_results[task_name] = {
                'v_t_scores': v_t_scores,
                'cycles_run': cycles_run,
                'coherence_scores': coherence_scores,
                'threshold': threshold,
                'max_cycles': max_cycles
            }

    # Plotting Metadata for all tasks
    for task_name, results in all_results.items():
        v_t_scores = results['v_t_scores']
        cycles_run = results['cycles_run']
        coherence_scores = results['coherence_scores']
        threshold = results['threshold']
        max_cycles = results['max_cycles']

        if v_t_scores:
            plt.figure(figsize=(10, 5))
            plt.hist(v_t_scores, bins=20, color='skyblue', edgecolor='black')
            plt.title(f'Distribution of $v_t$ Trigger Values ({task_name})')
            plt.xlabel('$v_t$ Trigger Value')
            plt.ylabel('Frequency')
            plt.axvline(threshold, color='red', linestyle='dashed', linewidth=1, label=f'Trigger Threshold ({threshold})')
            plt.legend()
            plt.grid(axis='y', alpha=0.75)
            plt.show()

        if cycles_run:
            plt.figure(figsize=(10, 5))
            plt.hist(cycles_run, bins=range(1, max_cycles + 2), color='lightcoral', edgecolor='black', align='left')
            plt.title(f'Distribution of Cycles Run when Confessional Triggered ({task_name})')
            plt.xlabel('Number of Cycles')
            plt.ylabel('Frequency')
            plt.xticks(range(1, max_cycles + 2))
            plt.grid(axis='y', alpha=0.75)
            plt.show()

        if coherence_scores:
            plt.figure(figsize=(10, 5))
            plt.hist(coherence_scores, bins=20, color='lightgreen', edgecolor='black')
            plt.title(f'Distribution of Average Final Coherence Scores when Confessional Triggered ({task_name})')
            plt.xlabel('Average Final Coherence Score')
            plt.ylabel('Frequency')
            plt.grid(axis='y', alpha=0.75)
            plt.show()
```

Modify the `truthfulqa_eval.py` script to accept a command-line argument for dataset selection, encapsulate dataset-specific logic in functions, and update the main execution block to call the appropriate function based on the argument.



```python
%%writefile truthfulqa_eval.py
import torch
import torch.nn as nn
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModel
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
import argparse # Import argparse

# Assuming cal.py is in the same directory and contains the module definitions
from cal import CAL_TRM_Hybrid, TinyConfessionalLayer

# 1. Define model parameters and device
d_model = 256
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 2. Instantiate the CAL_TRM_Hybrid model and move to device
model = CAL_TRM_Hybrid(d_model=d_model)
model.to(device)

# 3. Initialize Tokenizer and Embedding Model
print("Initializing tokenizer and embedding model...")
tokenizer = AutoTokenizer.from_pretrained('prajjwal1/bert-tiny')
embed_model = AutoModel.from_pretrained('prajjwal1/bert-tiny')
embed_model.to(device)
proj = nn.Linear(128, d_model)
proj.to(device)
print("Tokenizer and embedding model initialized.")

def load_and_select_dataset(dataset_name, num_examples=200):
    """Loads and selects a subset of the specified dataset."""
    print(f"Loading \"{dataset_name}\" dataset...")
    try:
        if dataset_name == "truthfulqa":
             dataset = load_dataset("truthful_qa", "generation")
        elif dataset_name in ["disambiguation_qa", "web_of_lies"]:
             # Attempt to load from tasksource/bigbench first
             try:
                  dataset = load_dataset("tasksource/bigbench", dataset_name)
             except Exception:
                  print(f"Could not load '{dataset_name}' from 'tasksource/bigbench'. Attempting direct load...")
                  dataset = load_dataset(dataset_name)
        else:
            raise ValueError(f"Unsupported dataset: {dataset_name}")

        split = "validation" if "validation" in dataset else ("test" if "test" in dataset else "train")
        evaluation_data = dataset[split]
        print(f"Loaded {len(evaluation_data)} examples from the '{split}' split.")

        evaluation_data = evaluation_data.select(range(min(num_examples, len(evaluation_data))))
        print(f"Selected {len(evaluation_data)} examples for evaluation.")
        return evaluation_data

    except Exception as e:
        print(f"Error loading dataset '{dataset_name}': {e}")
        return None


def evaluate_truthfulqa(data, model, tokenizer, embed_model, proj, device):
    """Evaluates the model on the TruthfulQA dataset."""
    correct_predictions = 0
    total_examples = 0
    all_v_t_scores = []
    all_cycles_run = []
    all_coherence_scores = []
    all_confessional_triggered_flags = []

    print("Starting evaluation on TruthfulQA...")

    prev_scratchpad_state = None

    for i, example in enumerate(data):
        question = example['question']
        correct_answers = example['correct_answers']
        incorrect_answers = example['incorrect_answers']

        if not correct_answers:
            continue

        total_examples += 1
        best_answer = correct_answers[0]
        input_text = f"Question: {question} Answer: {best_answer}"
        input_ids = tokenizer.encode(input_text, return_tensors='pt', padding=True, truncation=True).to(device)

        if input_ids.size(1) == 0:
            print(f"Skipping example {i} due to empty input after tokenization.")
            continue

        with torch.no_grad():
            bert_output = embed_model(input_ids)
            embeddings = bert_output.last_hidden_state
        projected_embeddings = proj(embeddings)

        with torch.no_grad():
            model_output, metadata, current_scratchpad_state = model(projected_embeddings, prev_z=prev_scratchpad_state, return_metadata=True)

        prev_scratchpad_state = None # Reset for next example in this loop structure

        model_output_pooled = torch.mean(model_output, dim=1)
        projected_embeddings_pooled = torch.mean(projected_embeddings, dim=1)
        cosine_score = F.cosine_similarity(model_output_pooled, projected_embeddings_pooled, dim=-1).item()

        if cosine_score > 0.95:
            correct_predictions += 1

        if 'v_t_trigger_values' in metadata:
            all_v_t_scores.extend(metadata['v_t_trigger_values'].tolist())

        all_confessional_triggered_flags.append(metadata.get('confessional_triggered', False))

        if metadata.get('confessional_triggered', False) and 'cal_metadata' in metadata:
             cal_metadata = metadata['cal_metadata']
             if 'cycles_run' in cal_metadata:
                  all_cycles_run.append(cal_metadata['cycles_run'])
             if 'coherence_score' in cal_metadata:
                  all_coherence_scores.append(cal_metadata['coherence_score'])

        if (i + 1) % 50 == 0:
            print(f"Processed {i + 1} examples. Current accuracy: {correct_predictions / total_examples if total_examples > 0 else 0:.4f}")

    accuracy = correct_predictions / total_examples if total_examples > 0 else 0
    triggered_count = sum(all_confessional_triggered_flags)
    trigger_rate = triggered_count / total_examples if total_examples > 0 else 0
    avg_cycles = np.mean(all_cycles_run) if all_cycles_run else 0
    avg_coherence = np.mean(all_coherence_scores) if all_coherence_scores else 0

    print("-" * 30)
    print("Evaluation complete for TruthfulQA.")
    print(f"Total examples evaluated: {total_examples}")
    print(f"Correct predictions (Cosine Sim > 0.95): {correct_predictions}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Trigger Rate (Confessional Triggered): {trigger_rate:.4f}")
    print(f"Average Cycles Run (when triggered): {avg_cycles:.2f}")
    print(f"Average Final Coherence (when triggered): {avg_coherence:.4f}")

    return all_v_t_scores, all_cycles_run, all_coherence_scores, model.threshold, model.cal_confessional.max_cycles


def evaluate_disambiguation_qa(data, model, tokenizer, embed_model, proj, device):
    """Evaluates the model on the disambiguation_qa dataset."""
    correct_predictions = 0
    total_examples = 0
    all_v_t_scores = []
    all_cycles_run = []
    all_coherence_scores = []
    all_confessional_triggered_flags = []

    print("Starting evaluation on disambiguation_qa...")

    prev_scratchpad_state = None

    for i, example in enumerate(data):
        input_text_base = example.get('inputs', '')
        options = example.get('multiple_choice_targets', [])
        correct_answer_index = -1
        if 'multiple_choice_scores' in example and 1 in example['multiple_choice_scores']:
            correct_answer_index = example['multiple_choice_scores'].index(1)

        if not options or correct_answer_index == -1 or correct_answer_index >= len(options):
            print(f"Skipping example {i} due to missing or invalid options/answer.")
            continue

        total_examples += 1
        input_texts = [f"{input_text_base} {option}" for option in options]
        correct_option_text = options[correct_answer_index]

        option_outputs = []
        option_v_t_scores = []
        option_cycles_run = []
        option_coherence_scores = []
        option_triggered_flags = []

        for input_text in input_texts:
            input_ids = tokenizer.encode(input_text, return_tensors='pt', padding=True, truncation=True).to(device)
            if input_ids.size(1) == 0: continue

            with torch.no_grad():
                bert_output = embed_model(input_ids)
                embeddings = bert_output.last_hidden_state
                if embeddings is None or embeddings.numel() == 0: continue
            projected_embeddings = proj(embeddings)

            with torch.no_grad():
                model_output, metadata, current_scratchpad_state = model(projected_embeddings, prev_z=prev_scratchpad_state, return_metadata=True)

            model_output_pooled = torch.mean(model_output, dim=1)
            option_outputs.append(model_output_pooled.squeeze(0).cpu().numpy())

            if 'v_t_trigger_values' in metadata:
                option_v_t_scores.extend(metadata['v_t_trigger_values'].tolist())
            option_triggered_flags.append(metadata.get('confessional_triggered', False))
            if metadata.get('confessional_triggered', False) and 'cal_metadata' in metadata:
                 cal_metadata = metadata['cal_metadata']
                 if 'cycles_run' in cal_metadata:
                      option_cycles_run.append(cal_metadata['cycles_run'])
                 if 'coherence_score' in cal_metadata:
                      option_coherence_scores.append(cal_metadata['coherence_score'])

        best_option_similarity = -float('inf')
        predicted_answer_index = -1

        for j, input_text in enumerate(input_texts):
            input_ids = tokenizer.encode(input_text, return_tensors='pt', padding=True, truncation=True).to(device)
            if input_ids.size(1) == 0: continue

            with torch.no_grad():
                bert_output = embed_model(input_ids)
                embeddings = bert_output.last_hidden_state
                if embeddings is None or embeddings.numel() == 0: continue
            projected_embeddings = proj(embeddings)
            projected_embeddings_pooled = torch.mean(projected_embeddings, dim=1).squeeze(0).cpu().numpy()

            if j < len(option_outputs):
                model_output_pooled_np = option_outputs[j]
                if np.linalg.norm(model_output_pooled_np) > 1e-8 and np.linalg.norm(projected_embeddings_pooled) > 1e-8:
                    similarity = np.dot(model_output_pooled_np, projected_embeddings_pooled) / (np.linalg.norm(model_output_pooled_np) * np.linalg.norm(projected_embeddings_pooled) + 1e-8)
                    if similarity > best_option_similarity:
                        best_option_similarity = similarity
                        predicted_answer_index = j
                else:
                    print(f"Warning: Zero vector encountered for similarity calculation in example {i}, option {j}.")

        if predicted_answer_index == correct_answer_index:
            correct_predictions += 1

        correct_input_text = f"{input_text_base} {correct_option_text}"
        correct_input_ids = tokenizer.encode(correct_input_text, return_tensors='pt', padding=True, truncation=True).to(device)
        if correct_input_ids.size(1) > 0:
            with torch.no_grad():
                correct_bert_output = embed_model(correct_input_ids)
                correct_embeddings = correct_bert_output.last_hidden_state
            if correct_embeddings is not None and correct_embeddings.numel() > 0:
                 correct_projected_embeddings = proj(correct_embeddings)
                 with torch.no_grad():
                      _, _, prev_scratchpad_state = model(correct_projected_embeddings, prev_z=prev_scratchpad_state, return_metadata=True)
            else:
                 prev_scratchpad_state = None
        else:
            prev_scratchpad_state = None

        all_v_t_scores.extend(option_v_t_scores)
        all_cycles_run.extend(option_cycles_run)
        all_coherence_scores.extend(option_coherence_scores)
        all_confessional_triggered_flags.append(any(option_triggered_flags))

        if (i + 1) % 50 == 0:
            print(f"Processed {i + 1} examples. Current accuracy: {correct_predictions / total_examples if total_examples > 0 else 0:.4f}")

    accuracy = correct_predictions / total_examples if total_examples > 0 else 0
    triggered_count = sum(all_confessional_triggered_flags)
    trigger_rate = triggered_count / total_examples if total_examples > 0 else 0
    avg_cycles = np.mean(all_cycles_run) if all_cycles_run else 0
    avg_coherence = np.mean(all_coherence_scores) if all_coherence_scores else 0

    print("-" * 30)
    print("Evaluation complete for disambiguation_qa.")
    print(f"Total examples evaluated: {total_examples}")
    print(f"Correct predictions (Disambiguation Task): {correct_predictions}")
    print(f"Accuracy (Disambiguation Task): {accuracy:.4f}")
    print(f"Trigger Rate (Confessional Triggered - across options): {trigger_rate:.4f}")
    print(f"Average Cycles Run (when triggered): {avg_cycles:.2f}")
    print(f"Average Final Coherence (when triggered): {avg_coherence:.4f}")

    return all_v_t_scores, all_cycles_run, all_coherence_scores, model.threshold, model.cal_confessional.max_cycles


def evaluate_web_of_lies(data, model, tokenizer, embed_model, proj, device):
    """Evaluates the model on the web_of_lies dataset."""
    correct_predictions = 0
    total_examples = 0
    all_v_t_scores = []
    all_cycles_run = []
    all_coherence_scores = []
    all_confessional_triggered_flags = []

    print("Starting evaluation on web_of_lies...")

    prev_scratchpad_state = None

    for i, example in enumerate(data):
        context = example.get('context', '')
        claim = example.get('claim', '')
        label = example.get('label', None) # 0 or 1

        if context is None or claim is None or label is None:
            print(f"Skipping example {i} due to missing context, claim, or label.")
            continue

        total_examples += 1
        input_text = f"Context: {context}\nClaim: {claim}"
        input_ids = tokenizer.encode(input_text, return_tensors='pt', padding=True, truncation=True).to(device)

        if input_ids.size(1) == 0:
            print(f"Warning: Empty input for example {i}.")
            continue

        with torch.no_grad():
            bert_output = embed_model(input_ids)
            embeddings = bert_output.last_hidden_state

        if embeddings is None or embeddings.numel() == 0:
            print(f"Warning: Empty embeddings for example {i}.")
            continue

        projected_embeddings = proj(embeddings)

        with torch.no_grad():
            model_output, metadata, current_scratchpad_state = model(projected_embeddings, prev_z=prev_scratchpad_state, return_metadata=True)

        prev_scratchpad_state = current_scratchpad_state # Pass state across examples

        mean_v_t = np.mean(metadata.get('v_t_trigger_values', [0]))

        if (label == 0 and mean_v_t > model.threshold) or \
           (label == 1 and mean_v_t <= model.threshold):
            correct_predictions += 1

        if 'v_t_trigger_values' in metadata:
            all_v_t_scores.extend(metadata['v_t_trigger_values'].tolist())

        all_confessional_triggered_flags.append(metadata.get('confessional_triggered', False))

        if metadata.get('confessional_triggered', False) and 'cal_metadata' in metadata:
             cal_metadata = metadata['cal_metadata']
             if 'cycles_run' in cal_metadata:
                  all_cycles_run.append(cal_metadata['cycles_run'])
             if 'coherence_score' in cal_metadata:
                  all_coherence_scores.append(cal_metadata['coherence_score'])

        if (i + 1) % 50 == 0:
            print(f"Processed {i + 1} examples. Current accuracy: {correct_predictions / total_examples if total_examples > 0 else 0:.4f}")

    accuracy = correct_predictions / total_examples if total_examples > 0 else 0
    triggered_count = sum(all_confessional_triggered_flags)
    trigger_rate = triggered_count / total_examples if total_examples > 0 else 0
    avg_cycles = np.mean(all_cycles_run) if all_cycles_run else 0
    avg_coherence = np.mean(all_coherence_scores) if all_coherence_scores else 0

    print("-" * 30)
    print("Evaluation complete for web_of_lies.")
    print(f"Total examples evaluated: {total_examples}")
    print(f"Correct predictions (Web of Lies Task): {correct_predictions}")
    print(f"Accuracy (Web of Lies Task): {accuracy:.4f}")
    print(f"Trigger Rate (Confessional Triggered): {trigger_rate:.4f}")
    print(f"Average Cycles Run (when triggered): {avg_cycles:.2f}")
    print(f"Average Final Coherence (when triggered): {avg_coherence:.4f}")

    return all_v_t_scores, all_cycles_run, all_coherence_scores, model.threshold, model.cal_confessional.max_cycles


def plot_metadata(all_v_t_scores, all_cycles_run, all_coherence_scores, trigger_threshold, max_cycles, dataset_name):
    """Plots the collected metadata."""
    if all_v_t_scores:
        plt.figure(figsize=(10, 5))
        plt.hist(all_v_t_scores, bins=20, color='skyblue', edgecolor='black')
        plt.title(f'Distribution of $v_t$ Trigger Values ({dataset_name} Task)')
        plt.xlabel('$v_t$ Trigger Value')
        plt.ylabel('Frequency')
        plt.axvline(trigger_threshold, color='red', linestyle='dashed', linewidth=1, label=f'Trigger Threshold ({trigger_threshold})')
        plt.legend()
        plt.grid(axis='y', alpha=0.75)
        plt.show()

    if all_cycles_run:
        plt.figure(figsize=(10, 5))
        plt.hist(all_cycles_run, bins=range(1, max_cycles + 2), color='lightcoral', edgecolor='black', align='left')
        plt.title(f'Distribution of Cycles Run (when Confessional Triggered - {dataset_name} Task)')
        plt.xlabel('Number of Cycles')
        plt.ylabel('Frequency')
        plt.xticks(range(1, max_cycles + 2))
        plt.grid(axis='y', alpha=0.75)
        plt.show()

    if all_coherence_scores:
        plt.figure(figsize=(10, 5))
        plt.hist(all_coherence_scores, bins=20, color='lightgreen', edgecolor='black')
        plt.title(f'Distribution of Average Final Coherence Scores (when Confessional Triggered - {dataset_name} Task)')
        plt.xlabel('Average Final Coherence Score')
        plt.ylabel('Frequency')
        plt.grid(axis='y', alpha=0.75)
        plt.show()


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Evaluate CAL-TRM on different datasets.")
    parser.add_argument("dataset", type=str, choices=["truthfulqa", "disambiguation_qa", "web_of_lies"],
                        help="Dataset to evaluate on (truthfulqa, disambiguation_qa, web_of_lies)")
    args = parser.parse_args()

    dataset_name = args.dataset

    evaluation_data = load_and_select_dataset(dataset_name)

    if evaluation_data is not None:
        if dataset_name == "truthfulqa":
            v_t_scores, cycles_run, coherence_scores, threshold, max_cycles = evaluate_truthfulqa(
                evaluation_data, model, tokenizer, embed_model, proj, device
            )
        elif dataset_name == "disambiguation_qa":
            v_t_scores, cycles_run, coherence_scores, threshold, max_cycles = evaluate_disambiguation_qa(
                evaluation_data, model, tokenizer, embed_model, proj, device
            )
        elif dataset_name == "web_of_lies":
            v_t_scores, cycles_run, coherence_scores, threshold, max_cycles = evaluate_web_of_lies(
                evaluation_data, model, tokenizer, embed_model, proj, device
            )

        plot_metadata(v_t_scores, cycles_run, coherence_scores, threshold, max_cycles, dataset_name)
```
